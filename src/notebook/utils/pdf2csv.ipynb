```python
df2 = pd.read_fwf('32_151_115.txt',pandas_options={'header': None})

df2 = df2['ELECTORAL ROLL 2020 S10 Karnataka'].str.split(':',expand=True)

df3 = df2.iloc[[0,4,5,24,30,39,40,51,52,53,54,55,56,57,58]]

df3.reset_index(inplace=True)

df3

df3[0].replace('Female Third Gender','Female',inplace=True)

df3.drop(2,axis=1,inplace=True)

df3.drop(9,axis=0,inplace=True)

df3.fillna('',inplace=True)

df3[2] = df3[0] + df3[1]

df4 = df3[[2]]

df4.loc[len(df4.index)+1, 2] = 'Third Gender'

df4

df_new = ['Assembly constituency','parliament constituency','Ward No','No. and Name of Polling Station','Male','Female','Third Gender','Total']

d = pd.DataFrame(df_new)

lists = []

df3

a = [0,2,4]

lists.append(df3[1][a])

lists

b = [6,8,11,12,14]

lists.append(df3[0][b])

lists

lists2 = []

for i in a:
    value = lists[0][i]
    lists2.append(value)
    print(value)

for i in b:
    value = lists[1][i]
    lists2.append(value)
    print(value)

f = pd.DataFrame(lists2)

dd = d+'/'+f

final = dd[0].str.split('/',expand=True)

final.columns = ['variable','details']

final
```
